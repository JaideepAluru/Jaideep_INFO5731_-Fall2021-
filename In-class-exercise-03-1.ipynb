Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd

t1 = []
url = "https://raw.githubusercontent.com/JaideepAluru/Jaideep_INFO5731_-SPRING2022-/main/sample.txt"
tf = urllib.request.urlopen(url)  #This function is used to decode the url and get the data from it it is a predefined function

for i in tf:
    dec_data = i.decode("utf-8").replace('\r\n', '')    #this statement is used to decode the data present in the url and convert it into standard format
    if dec_data: 
        t1.append(dec_data)       #appending the decoded data to a list
df = pd.DataFrame (t1, columns = ['Text'])   #converting the list to data frame using pamdas
print(df)


```

```python
#To find Number of Sentences in a data frame

df['Num of sentences'] = df['Text'].apply(lambda y: len(str(y).split(".")))
df
```

```python
#To find Number of words in a data frame

df['Num of Words'] = df['Text'].apply(lambda y: len(str(y).split(" ")))
df
```

```python
#To find Number of characters in data frame

df['Num of Char'] = df['Text'].str.len()
df
```

```python
#To find the average word length of the list


def awl(x):
    jai = x.split() #here the sentence is being splitted into parts and being assigned to variable jai
    if len(jai) != 0:  #here we are checking if the variable is empty or not
        return(sum(len(deep) for deep in jai)/len(jai))
    else:
        return None

df['Avg Word Length'] = df['Text'].apply(lambda z: awl(z))
df
```

```python
#To find the number of special characters present in the list

def spchar(x):
    red = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            red = red + 1
    return red

df['Num of spec char'] = df['Text'].apply(lambda y: spchar(y))
df
```

```python
#To find the number of numerics present in the list

df['Num of num'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()])) #here we are considering it to be a seperate numeric if it is seperated with a space using the split function
df
```

```python
#To find number of uppercase words in a list

df['Num of upper case words'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isupper()])) #using split function we are finding the words with completely upper case
df
```

```python
df.to_csv('data.csv',index=True)
```
